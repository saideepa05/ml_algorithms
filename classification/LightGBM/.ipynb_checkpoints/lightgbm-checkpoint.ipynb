```python
class Leaf:
    def __init__(self, indices, weight):
        self.indices = indices
        self.weight = weight
```

```python
class LightGBMTree:
    def __init__(self, max_leaves=6, lam=1.0):
        self.max_leaves = max_leaves
        self.lam = lam
        self.leaves = []
    def fit(self, X, grad, hess):
        n_samples = X.shape[0]
        G = np.sum(grad)
        H = np.sum(hess)
        root_weight = -G / (H + self.lam)
        self.leaves = [Leaf(np.arange(n_samples), root_weight)]
        for _ in range(self.max_leaves - 1):
            best_gain = -float("inf")
            best_split = None
            for leaf in self.leaves:
                idx = leaf.indices
                X_leaf = X[idx]
                grad_leaf = grad[idx]
                hess_leaf = hess[idx]
                G_total = np.sum(grad_leaf)
                H_total = np.sum(hess_leaf)
                for feature in range(X.shape[1]):
                    thresholds = np.unique(X_leaf[:, feature])
                    for threshold in thresholds:
                        left_mask = X_leaf[:, feature] <= threshold
                        right_mask = ~left_mask
                        if np.sum(left_mask) == 0 or np.sum(right_mask) == 0:
                            continue
                        G_L = np.sum(grad_leaf[left_mask])
                        H_L = np.sum(hess_leaf[left_mask])
                        G_R = G_total - G_L
                        H_R = H_total - H_L
                        gain = (
                            (G_L ** 2) / (H_L + self.lam)+ (G_R ** 2) / (H_R + self.lam)- (G_total ** 2) / (H_total + self.lam))
                        if gain > best_gain:
                            best_gain = gain
                            best_split = (leaf, feature, threshold, left_mask, right_mask)
            if best_split is None:
                break

            leaf, feature, threshold, left_mask, right_mask = best_split
            idx = leaf.indices
            self.leaves.remove(leaf)
            left_idx = idx[left_mask]
            right_idx = idx[right_mask]
            w_left = -np.sum(grad[left_idx]) / (np.sum(hess[left_idx]) + self.lam)
            w_right = -np.sum(grad[right_idx]) / (np.sum(hess[right_idx]) + self.lam)
            self.leaves.append(Leaf(left_idx, w_left))
            self.leaves.append(Leaf(right_idx, w_right))
    def predict(self, X):
        preds = np.zeros(X.shape[0])
        for leaf in self.leaves:
            preds[leaf.indices] = leaf.weight
        return preds
```

```python
class LightGBMClassifierScratch:
    def __init__(
        self,
        n_estimators=50,
        learning_rate=0.1,
        max_leaves=6,
        lam=1.0
    ):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_leaves = max_leaves
        self.lam = lam
        self.models = []
        self.init_pred = None
    def sigmoid(self, x):
        x = np.clip(x, -20, 20)
        return 1 / (1 + np.exp(-x))
    def fit(self, X, y):
        y = np.array(y)
        pos_ratio = np.clip(np.mean(y), 1e-6, 1 - 1e-6)
        self.init_pred = np.log(pos_ratio / (1 - pos_ratio))
        F = np.full(len(y), self.init_pred)
        for _ in range(self.n_estimators):
            p = self.sigmoid(F)
            grad = p - y
            hess = p * (1 - p)
            tree = LightGBMTree(max_leaves=self.max_leaves,lam=self.lam)
            tree.fit(X, grad, hess)
            update = tree.predict(X)
            F += self.learning_rate * update
            self.models.append(tree)

    def predict_proba(self, X):
        F = np.full(X.shape[0], self.init_pred)
        for tree in self.models:
            F += self.learning_rate * tree.predict(X)
        return self.sigmoid(F)

    def predict(self, X):
        probs = self.predict_proba(X)
        return (probs >= 0.5).astype(int)
```

```python
lgbm = LightGBMClassifierScratch(
    n_estimators=10,
    learning_rate=0.1,
    max_leaves=6,
    lam=1.0
)

lgbm.fit(X_train_scaled, y_train.values)
y_pred_lgbm = lgbm.predict(X_test_scaled)
y_prob_lgbm = lgbm.predict_proba(X_test_scaled)
```
