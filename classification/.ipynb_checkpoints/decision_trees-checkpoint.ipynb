## Decision Tree Helper Functions

```python
class DecisionTreeNode:
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, value=None):
        self.feature_index = feature_index      # which feature we split on
        self.threshold = threshold              # value to compare
        self.left = left                        # left subtree
        self.right = right                      # right subtree
        self.value = value                      # class label if leaf node
```

## Gini Impurity Calculation

```python
def gini_impurity(y):
    ##Calculate Gini impurity for a set of labels
    classes = np.unique(y)
    impurity = 1.0
    # print(f"Classes: {classes}")
    for cls in classes:
        p = np.sum(y == cls) / len(y)
        # print(f"Class {cls}: p = {p}")
        impurity -= p ** 2
        # print(f"Impurity after subtracting {p**2}: {impurity}")
    # print(f"Final Gini Impurity = {impurity}")
    return impurity
```

## Split the Dataset

```python
def split_dataset(X, y, feature_index, threshold):
    ###Split dataset into left and right nodes.
    left_idx = np.where(X[:, feature_index] <= threshold)
    right_idx = np.where(X[:, feature_index] > threshold)
    return X[left_idx], y[left_idx], X[right_idx], y[right_idx]
```

## Find the Best Split

```python
def best_split(X, y):
    best_feature, best_threshold = None, None
    best_gain = -1
    current_impurity = gini_impurity(y)
    n_features = X.shape[1]

    for feature in range(n_features):
        thresholds = np.unique(X[:, feature])
        for threshold in thresholds:
            X_left, y_left, X_right, y_right = split_dataset(X, y, feature, threshold)

            if len(y_left) == 0 or len(y_right) == 0:
                continue

            # Weighted impurity
            left_impurity = gini_impurity(y_left)
            right_impurity = gini_impurity(y_right)
            p_left = len(y_left) / len(y)
            p_right = len(y_right) / len(y)
            gain = current_impurity - (p_left * left_impurity + p_right * right_impurity)

            if gain > best_gain:
                best_feature = feature
                best_threshold = threshold
                best_gain = gain
    
    return best_feature, best_threshold
```

## Build the Decision Tree Recursively

```python
def build_tree(X, y, depth=0, max_depth=5, min_samples=2):
    # Stopping conditions → create leaf node
    if depth >= max_depth or len(y) < min_samples or len(np.unique(y)) == 1:
        leaf_value = np.bincount(y).argmax()
        return DecisionTreeNode(value=leaf_value)

    feature, threshold = best_split(X, y)

    # failed to find split → make leaf
    if feature is None:
        leaf_value = np.bincount(y).argmax()
        return DecisionTreeNode(value=leaf_value)

    # split dataset
    X_left, y_left, X_right, y_right = split_dataset(X, y, feature, threshold)

    # recursively build tree
    left_child = build_tree(X_left, y_left, depth + 1, max_depth, min_samples)
    right_child = build_tree(X_right, y_right, depth + 1, max_depth, min_samples)

    return DecisionTreeNode(feature_index=feature, threshold=threshold, left=left_child, right=right_child)
```

## Predicting for One Sample

```python
def predict_sample(node, x):
    if node.value is not None:
        return node.value

    if x[node.feature_index] <= node.threshold:
        return predict_sample(node.left, x)

    return predict_sample(node.right, x)
```

## Final Decision Tree Class

```python
class DecisionTreeScratch:
    def __init__(self, max_depth=5, min_samples=2):
        self.max_depth = max_depth
        self.min_samples = min_samples
        self.root = None

    def fit(self, X, y):
        self.root = build_tree(X, y, max_depth=self.max_depth, min_samples=self.min_samples)

    def predict(self, X):
        return np.array([predict_sample(self.root, sample) for sample in X])
```

```python
tree = DecisionTreeScratch(max_depth=5)
tree.fit(X_train_scaled, y_train.values)

y_pred_tree = tree.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc

print("Accuracy:", accuracy_score(y_test, y_pred_tree))
print("Precision:", precision_score(y_test, y_pred_tree))
print("Recall:", recall_score(y_test, y_pred_tree))
print("F1 Score:", f1_score(y_test, y_pred_tree))
```

```python
cm = confusion_matrix(y_test, y_pred_tree)
sns.heatmap(cm, annot=True, fmt="d", cmap="Oranges")
plt.title("Decision Tree - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
y_prob_tree = y_pred_tree  # decision tree returns 0 or 1

fpr, tpr, _ = roc_curve(y_test, y_prob_tree)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"Decision Tree (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Decision Tree")
plt.legend()
plt.show()
```
